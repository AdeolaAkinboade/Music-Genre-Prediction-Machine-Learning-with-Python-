```python
#first we import the data

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split# split data into training and testing
from sklearn.metrics import accuracy_score #will give  us accuracy of our test vs learing data
# joblib.dump(model, 'music_recommendations.joblib') stores trained model in file

music_data = pd.read_csv(r'C:\Users\Jennifer\Desktop\music.csv')
```

```python
music_data #check to see if the data has run correctly with no errors
```

```python
#clean the data
#means removing duplicates, null values,delete columns,split dataset into input and output
#going to predict what music 21 year oldwould probably be into

X=music_data.drop(columns=['genre']) #doimg the music_data file plis tab, we can see the 'drop' attribute parameters and that 
                                  # will show us what to run and specify for code execution
                                 # this will allow us to have a view of data set withou 'genre'
                                  #Capital X to represent the expression music_data. drop()
```

```python
X
```

```python
#MAKING OUR OUTPUT SET. THE TOP ON WAS THE IMPUT

y= music_data['genre']
X_train, X_test, y_train , y_test =train_test_split(X, y, test_size= 0.2) #gave 20% of data to testing
```

```python
y # this shows the predictions (outputs ONLY)
```

```python
#Predicting Model using Algorhitum
#from sklearn
 
```

```python
model= DecisionTreeClassifier()
model.fit(X_train,y_train) #takes input set and output set, passing only training data not test part
predictions = model.predict(X_test)# it takes a two dimensional array
                            #put 21,1 like new record in the table
                            #another input set 22,0 (0 represents female and 1 male)
                            #asked model to make 2 predictions at once
                            #predictions is variable where it's stored
```

```python
predictions
```

```python
#ACCURACY MEASURE OF MODEL( TESTING PART, OTHER PART WAS TRAINING)

score= accuracy_score(y_test,predictions) #between 0-1
                                          #y testoutput vs prediction output
```
